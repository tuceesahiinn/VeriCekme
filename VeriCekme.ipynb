```python
liste = []

a=1
while a<=45 :

    headers_param={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"}
    r = requests.get("https://www.hepsiburada.com/akilli-saatler-c-60003676?sayfa="+str(a)+"",headers= headers_param)
    soup = BeautifulSoup(r.content,"lxml") 

    st1 = soup.find("div",attrs={"class":"contain-lg-3 contain-md-3 contain-sm-3 fluid with-bottom-border"})
    st2 = st1.find("ul",attrs={"data-bind":"css: {grid: isGridSelected, list: isListSelected}"})
    st3 = st2.find_all("li",attrs={"data-index":"1"})



    for detaylar in st3:
        link_sonu = detaylar.a.get("href")
        link_basi = "https://www.hepsiburada.com"
        link = link_basi + link_sonu
        print(link)
        
        r1 = requests.get(link,headers= headers_param)
        soup1 = BeautifulSoup(r1.content,"lxml")

        try:
            yeni_fiyat = soup1.find("span",attrs = {"data-bind":"text: product().currentListing.currentPriceBeforePoint + ',' + product().currentListing.currentPriceAfterPoint"}).text.strip()
            print(yeni_fiyat)
        except:
            yeni_fiyat=np.nan
        try:
            orijinal_fiyat = soup1.find("del",attrs={"data-bind":"markupText:'originalPrice', css: {'active' : hasDiscountPrice}"}).text
            print(orijinal_fiyat)
        except:
            orijinal_fiyat=np.nan
        try:
            indirim = soup1.find("span",attrs={"data-bind":"markupText: 'discountRate'"}).text
            print(indirim)
        except:
            indirim=np.nan
        try:
            puan = soup1.find("span",attrs={"class":"rating-star"}).text.strip()
            print(puan)
        except:
            puan=np.nan
        try:
            degerlendirme = soup1.find("div",attrs={"id":"comments-container"}).text.strip()
            print(degerlendirme)
        except:
            degerlendirme=np.nan
        try:
            urun_adi = soup1.find("span",attrs={"class":"product-name"}).text
            print(urun_adi)
        except:
            urun_adi= np.nan
        try:
            urun_marka = soup1.find("span",attrs={"class":"brand-name"}).text.strip()
            print(urun_marka)
        except:
            urun_marka= np.nan
        try:
            urun_fotograf = soup1.img.get("src")
            print(urun_fotograf)
        except:
            urun_fotograf= np.nan

        liste.append([urun_adi,urun_marka,link,urun_fotograf,yeni_fiyat,indirim,orijinal_fiyat,puan,degerlendirme])
    
    a=a+1
```

```python
df = pd.DataFrame(liste)
df.columns = ["Ürünün Adı","Marka","Link","Fotoğraf","Yeni Fiyat","İndirim","Orijinal Fiyat","Puan","Değerlendirme"]
```

```python
df.to_excel("AkilliSaatler.xlsx")
```
